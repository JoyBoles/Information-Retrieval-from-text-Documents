### **Term Document Matrix**

```python
Terms = set()
```

```python
for i in range(514):
    Terms=Terms.union({term for term in s_stem[i]})
```

```python
Terms
```

```python
doc_term_matrix = {}
```

```python
for term in Terms:
    doc_term_matrix[term] = []
    for i in range(514):
        if term in s_stem[i]: 
            doc_term_matrix[term].append(1)
        else: 
            doc_term_matrix[term].append(0)
```

**Queries**

```python
import numpy as np
```

```python
query = input("Enter your query\n")
```

```python
def preprocess(query):
    tokens_ = word_tokenize(query)
    sw_rem_ = [ word for word in tokens_ if not word.lower() in stop_words ]
    res = [word for word in sw_rem_ if word not in punctuations]
    stem = [ss.stem(word) for word in res]
    return stem
```

```python
query_list = preprocess(query)
print(query_list)
v = np.ones((514,), dtype = np.int8)
```

```python
for word in query_list:
    print(word)
    #print(doc_term_matrix[word])
    arr = np.array(doc_term_matrix[word])
    v = v & arr

#print(f'v = {v}')
print(np.argwhere(v == 1).reshape(1,-1))
```

```python
v = np.zeros((514,), dtype = np.int8)
for word in query_list:
    print(word)
    #print(doc_term_matrix[word])
    arr = np.array(doc_term_matrix[word])
    v = v | arr
    
print(np.argwhere(v == 1).reshape(1,-1))
```

### **TF-IDF**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vectorizer = TfidfVectorizer(stop_words = list(stop_words))
```

```python
tfidf = vectorizer.fit_transform(docs)
```

**Calculate similarities between documents**

```python
similarities = np.zeros((514,514))
```

```python
for i in range(514):
    for j in range(514):
        similarities[i,j] = cosine_similarity(tfidf[i],tfidf[j])
```

```python
similarities_df = pd.DataFrame(similarities)
```

```python
similarities_df
```

**Query Handling**

```python
query = input("What are you searching for?")
```

```python
query_tfidf = vectorizer.transform([query])
```

```python
cosineSimilarities = cosine_similarity(query_tfidf, tfidf).flatten()
```

```python
cosineSimilarities
```

```python
top_res = cosineSimilarities[cosineSimilarities > 0.1]
```

```python
top_res
```

```python
rankings = np.argsort(cosineSimilarities)[::-1]
```
